```python
pbar = None


def show_progress(block_num, block_size, total_size):
    global pbar
    if pbar is None:
        pbar = progressbar.ProgressBar(maxval=total_size)
        pbar.start()

    downloaded = block_num * block_size
    if downloaded < total_size:
        pbar.update(downloaded)
    else:
        pbar.finish()
        pbar = None
```

```python
pages = ['lifestyle', 'nature', 'animal', 'food', 'transport', 'sky', 'sea', 'beach', 'water', 'flower', 'snow',
         'rain', 'fire', 'forest', 'sunset', 'sun', 'smoke', 'night', 'space', 'earth', 'cloud', 'people', 'woman',
         'man', 'couple', 'baby', 'business', 'sport', 'dance', 'music', 'house', 'love', 'family', 'party', 'fasion',
         'pet', 'dog', 'cat', 'bird', 'fish', 'wildlife', 'reptile', 'safari', 'zoo', 'kitten', 'drink', 'coffee',
         'tea', 'fruit', 'restaurant', 'beer', 'cocktail', 'salad', 'eating', 'vegetable', 'fast-food', 'car',
         'street', 'city', 'traffic', 'road', 'truck', 'airplane', 'taxi', 'bicycle', 'motorcycle', 'drive']
```

```python
count = 0
for k in pages:
    print(k)
    url = f"https://mixkit.co/free-stock-video/{k}/?page={1}"
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")
        num_pages = int(soup.find_all("a", class_="pagination__link")[-1].text)
    except:
        continue

    for i in tqdm(range(1, num_pages+1)):
        url = f"https://mixkit.co/free-stock-video/{k}/?page={i}"
        response = requests.get(url)

        soup = BeautifulSoup(response.text, "html.parser")

        # Find all the video thumbnail elements on the page
        video_thumbnails = soup.find_all("div", class_="item-grid-card")
        # print(video_thumbnails)
        # Loop through each video thumbnail and extract its URL and description
        for thumbnail in video_thumbnails:
            try:
                video_url = thumbnail.find('video', class_='item-grid-video-player__video')["src"]
                video_description = thumbnail.find("p", class_="item-grid-card__description").text.strip()
                if len(df[df['path'] == video_url]) == 0:
                    count += 1
                    df.loc[count-1, 'path'] = video_url
                    df.loc[count-1, 'ann'] = video_description
                    df.loc[count-1, 'class'] = k
            except AttributeError:
                pass
```

```python
df = df.drop_duplicates()
```

```python
df['name'] = df['path'].str.split('/').str[-1]
```

# translate caps

```python
import pandas as pd
import json
```

```python
df = pd.DataFrame({'id':[], 'category_id':[], 'question': [], 'answer': [], 'video_id': []})
```

```python
with open('MSRVTT-QA/val_qa.json') as file:
    data = json.load(file)
```

```python
data
```

```python
len(data)
```

```python
for i, d in enumerate(data):
    df.loc[i, 'id'] = d['id']
    df.loc[i, 'category_id'] = d['category_id']
    df.loc[i, 'answer'] = d['answer']
    df.loc[i, 'question'] = d['question']
    df.loc[i, 'video_id'] = d['video_id']
```

```python
df
```

```python
from easygoogletranslate import EasyGoogleTranslate
from time import sleep
```

```python
translator = EasyGoogleTranslate(
    source_language='en',
    target_language='ru',
    timeout=7
)
translator.translate('This is an example')
```

```python
for i in tqdm(df.index):
    try:
        df.loc[i, 'question_ru'] = translator.translate(df.loc[i, 'question'])
        df.loc[i, 'answer_ru'] = translator.translate(df.loc[i, 'answer'])
    except:
        translator = EasyGoogleTranslate(
        source_language='en',
        target_language='ru',
        timeout=7)
        df.loc[i, 'question_ru'] = translator.translate(df.loc[i, 'question'])
        df.loc[i, 'answer_ru'] = translator.translate(df.loc[i, 'answer'])
```

```python
df.to_csv('ru-msrvtt-qa.casv')
```
