```python
from easygoogletranslate import EasyGoogleTranslate
from tqdm import tqdm
import numpy as np
```

```python
translator = EasyGoogleTranslate(
    source_language='en',
    target_language='ru',
    timeout=7
)
translator.translate('is the no.3 athlete playing indoor')
```

```python
df = activity_df_val
```

```python
for i in tqdm(df.index):
    if not df.loc[i, 'question_ru'].__class__ == str:
        try:
            df.loc[i, 'question_ru'] = translator.translate(df.loc[i, 'question'])
            df.loc[i, 'answer_ru'] = translator.translate(df.loc[i, 'answer'])
        except:
            translator = EasyGoogleTranslate(
            source_language='en',
            target_language='ru',
            timeout=7)
            df.loc[i, 'question_ru'] = translator.translate(df.loc[i, 'question'])
            df.loc[i, 'answer_ru'] = translator.translate(df.loc[i, 'answer'])
```

```python
df.to_csv('ru_activity_val.csv')
```
